```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv(
    '../data/daily_barbacena.csv',
    sep=',',
    parse_dates=['Timestamp'],
    index_col=['Timestamp'])
```

```python
plt.figure(figsize=(16, 6))

heatmap = sns.heatmap(dataframe.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(dataframe.corr()[['RADIATION']].sort_values(by='RADIATION', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with Radiation', fontdict={'fontsize':18}, pad=16);
```
